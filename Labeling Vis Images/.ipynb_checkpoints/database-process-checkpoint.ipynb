## generate data for not string value and x,y,z

```python
import duckdb
import random
import math

# Connect to the database
con = duckdb.connect("publications_princeton.db")

# Add new columns to the figure_property table if they do not exist
con.execute("ALTER TABLE figure_property ADD COLUMN IF NOT EXISTS xPos INTEGER")
con.execute("ALTER TABLE figure_property ADD COLUMN IF NOT EXISTS yPos INTEGER")
con.execute("ALTER TABLE figure_property ADD COLUMN IF NOT EXISTS zPos INTEGER")
con.execute("ALTER TABLE figure_property ADD COLUMN IF NOT EXISTS score INTEGER")  # Add score column

# Generate random data for the existing tables
def generate_random_string():
    return ''.join(random.choices('abcdefghijklmnopqrstuvwxyzABCDEFGHIJKLMNOPQRSTUVWXYZ', k=8))

def generate_random_values(row_count):
    data = []
    for _ in range(row_count):
        int_value = random.randint(0, 44)
        string_value = generate_random_string()
        name = f"name-{int_value}"
        x_base = (int_value / math.pi) * (math.sin(int_value)) + random.uniform(-10, 10)
        y_base = (int_value / math.pi) * (math.cos(int_value)) + random.uniform(-10, 10)
        xPos = int(x_base * 100)
        yPos = int(y_base * 100)
        zPos = int(random.random() * 100)
        score = random.randint(1, 99999)  # Generate random score
        data.append((name, int_value, string_value, xPos, yPos, zPos, score))
    return data

# Fetch the figure ids from the figure table
figure_ids = con.execute("SELECT id FROM figure").fetchall()

# Generate random data
random_data = generate_random_values(len(figure_ids))

# Insert data into the figure_property table
insert_query = """
INSERT INTO figure_property (figure_id, name, int_value, string_value, xPos, yPos, zPos, score)
VALUES (?, ?, ?, ?, ?, ?, ?, ?)
"""

# Insert each row into the figure_property table
for figure_id, data in zip(figure_ids, random_data):
    con.execute(insert_query, (figure_id[0], *data))

# Commit the changes
con.commit()
```

# Code using labels and socre to create x,y,z cordinates 

```python
import duckdb
import math

# Connect to the database
con = duckdb.connect("publications_princeton-chart.db")

# Fetch data from the table
query = "SELECT name, int_value, string_value, figure_id, score FROM figure_property;"
data = con.execute(query).fetchdf()

# Sort data by int_value and then by score descending
data.sort_values(by=['int_value', 'score'], ascending=[True, False], inplace=True)

# Variables for grid management
spacing = 15  # Updated spacing between items within a group
gap = 25
max_dimension = 1120

# Placeholder for tracking group placements (x_start, z_start, x_end, z_end)
placed_groups = []

def find_space_for_group(side_length):
    # Calculate the physical dimensions of the group
    group_width = side_length * spacing
    # Attempt to find space on the 2000x2000 surface
    for z in range(-max_dimension, max_dimension - group_width, spacing + gap):
        for x in range(-max_dimension, max_dimension - group_width, spacing + gap):
            fits = True
            # Check if the square fits into the space without overlapping
            for (px, pz, px_end, pz_end) in placed_groups:
                if not (x + group_width <= px - gap or x - gap >= px_end or 
                        z + group_width <= pz - gap or z - gap >= pz_end):
                    fits = False
                    break
            if fits:
                return (x, z)
    return None  # If no space is found

# Process each group
for int_value, group in data.groupby('int_value'):
    n_items = len(group)
    side_length = math.ceil(math.sqrt(n_items))
    
    # Find space for this group with updated dimensions
    space = find_space_for_group(side_length)
    if not space:
        raise Exception("Unable to find space for group with int_value: {}".format(int_value))
    
    x_start, z_start = space
    x_end = x_start + side_length * spacing
    z_end = z_start + side_length * spacing
    placed_groups.append((x_start, z_start, x_end, z_end))

    # Calculate positions and y positions
    positions = [(x_start + (i % side_length) * spacing, z_start + (i // side_length) * spacing) for i in range(n_items)]
    max_score = group['score'].max()
    min_score = group['score'].min()
    y_positions = [int(100 * 1.5 * (score - min_score) / (max_score - min_score) if max_score != min_score else 150) for score in group['score']]

    # Update database
    for (x, z), y, figure_id in zip(positions, y_positions, group['figure_id']):
        con.execute("UPDATE figure_property SET xPos = ?, yPos = ?, zPos = ? WHERE figure_id = ?", (x, y, z, figure_id))

# Commit changes to the database
con.commit()
con.close()
```

```python
import duckdb

# Connect to the database
con = duckdb.connect("publications_princeton-chart-3.db ")

# Create a temporary table with row numbers
con.execute("""
    CREATE TEMPORARY TABLE temp_figure AS
    SELECT ROW_NUMBER() OVER (ORDER BY id) AS new_id, id
    FROM figure
""")

# Update the original table by joining it with the temporary table on the old id
con.execute("""
    UPDATE figure
    SET id = temp_figure.new_id 
    FROM temp_figure
    WHERE figure.id = temp_figure.id
""")

# Drop the temporary table
con.execute("DROP TABLE temp_figure")

# Commit changes and close the connection
con.close()
```

```python
import os
import requests
from PIL import Image
from io import BytesIO
import pandas as pd

# Function to download and save the image
def download_and_save_image(url, file_path):
    try:
        response = requests.get(url)
        response.raise_for_status()  # Raise an error for bad status codes
        img = Image.open(BytesIO(response.content))
        img = img.resize((32, 32))
        img.save(file_path)
    except requests.exceptions.RequestException as e:
        print(f"Error downloading {url}: {e}")
    except Exception as e:
        print(f"Error processing {url}: {e}")

# Function to create the required directories
def create_directories(base_dir, n_folders):
    for i in range(1, n_folders + 1):
        folder_path = os.path.join(base_dir, f"figures_{i:02d}")
        os.makedirs(folder_path, exist_ok=True)
```

```python
# Directory where images will be saved
base_dir = 'figures'
n_folders = (len(df) - 1) // 4096 + 1
create_directories(base_dir, n_folders)

# Download and save images
for index, row in df.iterrows():
    id_x = row['id_x']
    url = row['server_path']
    folder_number = (id_x) // 4096 + 1
    folder_path = os.path.join(base_dir, f"figures_{folder_number:02d}")
    file_name = f"figure_{id_x:05d}.jpg"
    file_path = os.path.join(folder_path, file_name)
    download_and_save_image(url, file_path)

print("Download and processing complete.")
```

```python
import duckdb
import random
import math

# Connect to DuckDB database
con = duckdb.connect("publications_princeto.db")

# Check if the columns xPos, yPos, and zPos exist in the figure_property table
columns = con.execute("PRAGMA table_info(figure_property)").fetchall()
column_names = [column[1] for column in columns]

# Add columns xPos, yPos, and zPos if they do not exist
if 'xPos' not in column_names:
    con.execute("ALTER TABLE figure_property ADD COLUMN xPos INTEGER DEFAULT 0")
if 'yPos' not in column_names:
    con.execute("ALTER TABLE figure_property ADD COLUMN yPos INTEGER DEFAULT 0")
if 'zPos' not in column_names:
    con.execute("ALTER TABLE figure_property ADD COLUMN zPos INTEGER DEFAULT 0")

# Fetch all rows from the figure table
figure_rows = con.execute("SELECT rowid, id FROM figure").fetchall()

# Update figure.id with the index (row number)
for index, (rowid, old_id) in enumerate(figure_rows):
    con.execute("UPDATE figure SET id = ? WHERE rowid = ?", (index, rowid))

# Define possible values for int_value and string_value
int_values = [random.randint(0, 44) for _ in range(40)]
string_values = [f'string_{i}' for i in range(40)]

# Clear the figure_property table
con.execute("DELETE FROM figure_property")

# Dictionary to store generated x and y values for each int_value
value_positions = {}

# Function to generate nearby values
def generate_nearby_values(base_value, variance):
    return base_value + random.uniform(-variance, variance)

# Insert new entries into the figure_property table for each updated figure.id
for index, (rowid, old_id) in enumerate(figure_rows):
    # Generate random values for figure_property
    int_value = random.choice(int_values)

    if int_value not in value_positions:
        x_base = (int_value/math.pi)*(math.sin(int_value)) + random.uniform(-10, 10)
        y_base = (int_value/math.pi)*(math.cos(int_value)) + random.uniform(-10, 10)
        value_positions[int_value] = (x_base, y_base)
    else:
        x_base, y_base = value_positions[int_value]

    xP = generate_nearby_values(x_base, 2)
    yP = generate_nearby_values(y_base, 2)
    zP = (int_value + random.uniform(-10, 10)) / (int_value if int_value != 0 else 1)  # Avoid division by zero

    xPos = int(xP * 100)
    yPos = int(yP * 100)
    zPos = int(zP * 100)

    string_value = random.choice(string_values)
    name = f'name_{random.randint(1, 100)}'

    # Insert new figure_property entry
    con.execute(
        "INSERT INTO figure_property (name, int_value, string_value, xPos, yPos, zPos, figure_id) VALUES (?, ?, ?, ?, ?, ?, ?)",
        (name, int_value, string_value, xPos, yPos, zPos, index)
    )

# Commit the changes to the database
con.commit()

# Close the connection to the temporary database file
con.close()

# Reconnect to your DuckDB database
con = duckdb.connect("publications_princeton_int.db")

# Save the updated database as a new file
# con.execute("EXPORT DATABASE 'publications_princeton_pure.db'")

# Commit and close the connection
con.commit()
con.close()

print("Database updated successfully and saved as 'publications_princeton_updated.db'.")
```

```python
import duckdb

# Connect to the database
con = duckdb.connect("publications_princeton-6.db")

# Ensure the 'charts' table exists
con.execute("""
CREATE TABLE IF NOT EXISTS charts (
    id INTEGER,
    chart VARCHAR,
    xPos DOUBLE,
    yPos DOUBLE,
    zPos DOUBLE
)
""")

# Retrieve chart types list
chart_types = [
    "Violin", "Density", "Histogram", "Boxplot", "Ridgeline", "Scatter", "Heatmap", "Correlogram", "Bubble",
    "Connected scatter", "Density 2d", "Barplot", "Spider / Radar", "Wordcloud", "Parallel", "Lollipop",
    "Circular Barplot", "Treemap", "Venn diagram", "Doughnut", "Pie chart", "Dendrogram", "Circular packing",
    "Sunburst", "Line plot", "Area", "Stacked area", "Streamchart", "Map", "Choropleth", "Hexbin map", "Cartogram",
    "Connection", "Bubble map", "Chord diagram", "Network", "Sankey", "Arc diagram", "Edge bundling", "Complex",
    "Scientific Viz", "Other-1", "Other-2", "Other-3", "Other-4"
]

# Calculate average positions
result = con.execute("""
SELECT
    int_value,
    AVG(xPos) AS xPos,
    AVG(yPos) AS yPos,
    AVG(zPos) AS zPos
FROM figure_property
GROUP BY int_value
""").fetchall()

# Insert the results with the corresponding chart type into 'charts' table
for index, (int_value, avg_x, avg_y, avg_z) in enumerate(result):
    chart_type = chart_types[index % len(chart_types)]  # Cyclic assignment of chart types
    con.execute("""
    INSERT INTO charts (id, chart, xPos, yPos, zPos)
    VALUES (?, ?, ?, ?, ?)
    """, (int_value, chart_type, avg_x, avg_y, avg_z))

# Commit the transaction to save changes to the database
con.commit()

# Close the connection
con.close()
```

```python
import duckdb

# Connect to the database
con = duckdb.connect("publications_princeton-chart-5.db", read_only=False)

try:
    # Create a new table with updated figure_id
    con.execute("""
        CREATE TABLE updated_figure_property AS
        SELECT
            row_number() OVER (ORDER BY figure_id) AS new_figure_id,
            name,
            int_value,
            string_value,
            xPos,
            yPos,
            zPos,
            score
        FROM figure_property
    """)

    # Drop the old table
    con.execute("DROP TABLE figure_property")

    # Rename the new table to the original name
    con.execute("ALTER TABLE updated_figure_property RENAME TO figure_property")

    # Update the column name to figure_id
    con.execute("ALTER TABLE figure_property RENAME COLUMN new_figure_id TO figure_id")

    # Commit the changes to the database
    con.commit()
finally:
    # Close the connection
    con.close()

# Verify the updates (optional, for testing in the same script)
con = duckdb.connect("publications_princeton-chart-5.db")
updated_df = con.execute("SELECT * FROM figure_property").df()
print(updated_df)
```
