```python
# import libraries
import json
import pandas as pd
import pprintpp as pp
```

```python
# load file
current_file = 'urls_2019.json'
j = json.load(open(f'{current_file}'))

# total number of objects
print("works: ", len(j))

# load file
current_file_authors = 'authors_urls_2019.json'
j_authors = json.load(open(f'{current_file_authors}'))

# total number of objects
print("authors: ", len(j_authors))
```
