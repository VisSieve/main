```python
#Just some bs4 code to digest the name contents from the html pages
import re
import tqdm
names = []
for html in tqdm.tqdm(htmls):
    soup = bs4.BeautifulSoup(html.read_text())
    h3s = soup.find_all("h3")
    if len(h3s) ==0:
        # print("\t",html)
        pass
    else:
        name = h3s[0].text
        match = re.search("(.*), (.*)",name)
        last = match.group(1)
        first = match.group(2)
        middle = ""
        if " " in first:
            middle = first.split(" ")[1]
            first= first.split(" ")[0]
        names.append({"first":first,"middle":middle,"last":last})
        


names

    
```

```python
import requests as rq
import random
import time
```

```python
# demonstration of using openalex from python with the requests package
def open_alex_request(name):
    time.sleep(random.random()*.3)
    first = name["first"]
    middle = name["middle"]
    last = name["last"]
    headers = {"mailto":"baylyd@arizona.edu"}
    res = rq.get(f"https://api.openalex.org/authors?search={first} {middle} {last}",headers=headers)
    return res
```

```python
results = []
for name in tqdm.tqdm(names):
    res = open_alex_request(name)
    results.append({"name":name,"result":res.json()})
```

tackle the case where there's more than one person LATER

```python
# more open alex requests work, gather the works for this particular author
urls = []
works_collected = []
for res in tqdm.tqdm(good_res):
    person = res["result"]["results"][0]
    
    works = person["works_api_url"]
    
    headers = {"mailto":"baylyd@arizona.edu"}
    # get their works
    time.sleep(random.random()*.3)
    works_res = rq.get(works,headers=headers)
    
    works_cont = works_res.json()
    # consider case where they have more than 25 hits and we need to page
    works_collected.append({"name":res["name"],"works":works_cont})
    #loop here
    for work in works_cont["results"]:
        oa = work["open_access"]
        
        if oa["is_oa"]:
            url = oa["oa_url"]
            urls.append(url)
```
